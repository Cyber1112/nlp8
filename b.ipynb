```python
import pandas as pd
import spacy
import re
```

```python
def TextCleaning(document):
    text=re.sub(pattern='<.*?>',repl='',string=document)
    text=re.sub(pattern='[^a-zA-Z0-9\s]',repl='',string=text)
    text=re.sub(pattern='\s+',repl=' ',string=text)
    text=str(text).lower()
    tokens=[token for token in text.split(' ') if token != '']
    return ' '.join(tokens)
```

```python
df = pd.read_csv("./data/imbd.csv")
df["review"] = df["review"].apply(TextCleaning)

adj = {}
```

```python
nlp = spacy.load('en_core_web_md')

for review, sentiment in zip(df['review'], df['sentiment']):
    doc = nlp(review)

    for token in doc:
        if token.pos_ == 'ADJ':
            lemma = token.lemma_
            if lemma not in adj:
                adj[lemma] = {'positive': 0, 'negative': 0}

            adj[lemma][sentiment] += 1
```
