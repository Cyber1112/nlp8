```python
import os
import shutil
import zipfile
import pandas as pd
import re
import spacy
```

```python
def prepare_files(data_path):
    file_extension = ".txt"
    # Initialize an empty list to store the file paths
    file_paths = []
    # Use os.walk to traverse through the directory and subdirectories
    for dirpath, dirnames, filenames in os.walk(data_path):
        for filename in filenames:
            if filename.endswith(file_extension):
                # Construct the full file path
                file_path = os.path.join(dirpath, filename)
                # Append the file path to the list
                file_paths.append(file_path)
    return file_paths
```

```python
named_entities = {}

def save_named_entities(text, tag):
    nlp = spacy.load("en_core_web_sm")
    doc = nlp(text)

    for entity in doc.ents:
        entity_text = entity.text

        if entity_text in named_entities:
            if tag in named_entities[entity_text]:
                named_entities[entity_text][tag] += 1
            else:
                named_entities[entity_text][tag] = 1
        else:
            named_entities[entity_text] = {tag : 1}
```

```python
files = prepare_files("./bbc/business")
for file_path in files:
    with open(file_path, "r", encoding='unicode_escape') as text_file:
        # Read the text from the file
        text = text_file.read()
        save_named_entities(text, "business")
```

```python
entertainment_files = prepare_files("./bbc/entertainment")
for file_path in entertainment_files:
    with open(file_path, "r", encoding='unicode_escape') as text_file:
        text = text_file.read()
        save_named_entities(text, "entertainment")
```
